```python
import pandas as pd

df = pd.read_csv("iris.data", header=None, names=["sepal_length", "sepal_width", "petal_length", "petal_width", "class"])
print(df)  
```

```python
print("Número de linhas e colunas:", df.shape)

print("\nInformações:")
df.info()

print("\nEstatísticas:")
print(df.describe())
```

#### Exercício 3: Gráficos
Gere os seguintes gráficos usando a biblioteca matplotlib ou seaborn:
- Um histograma para cada uma das colunas numéricas.
- Um boxplot para as colunas numéricas, segmentadas por espécie.
- Um scatter plot (diagrama de dispersão) relacionando pares de colunas.

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].hist()
plt.suptitle("Histograma das colunas numéricas")
plt.show()
```

```python
nomes_colunas = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
colunas_numericas = nomes_colunas[:-1]  

for col in colunas_numericas:
  df.boxplot(column=col, by='class')
  plt.title(f"Boxplot da coluna {col}")
  plt.xlabel("Espécie")
  plt.ylabel(col)
  plt.show()
```

```python
import seaborn as sns
sns.pairplot(df, hue="class")
plt.suptitle("Diagrama de Dispersão em pares", fontsize=16)
plt.show()
```

#### Exercício 4: Funções para modularização
Transforme cada uma das etapas anteriores em funções reutilizáveis, como:
- load_dataset(file_path) para carregar o dataset.
- dataset_summary(df) para exibir informações e estatísticas básicas.
- plot_histogram(df, column) para criar um histograma.

```python
def load_dataset(file_path):
  """
  Carrega o dataset Iris a partir do arquivo especificado e retorna um DataFrame.
  
  Parâmetros:
    file_path (str): Caminho para o arquivo 'iris.data'.
  
  Retorna:
    df (DataFrame): O dataset Iris com nomes de coluna definidos.
  """
  nomes_colunas = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
  df = pd.read_csv(file_path, header=None, names=nomes_colunas)
  return df
```

```python
def dataset_summary(df):
  """
  Exibe informações e estatísticas do DataFrame.
  
  Parâmetros:
    df (DataFrame): Dataset base.
  """
  print("Número de linhas e colunas:", df.shape)
  
  print("\nInformações do DataFrame:")
  df.info()
  
  print("\nEstatísticas descritivas:")
  print(df.describe())
```

```python
def plot_histogram(df, column):
  """
  Plota um histograma para uma coluna específica do DataFrame.
  
  Parâmetros:
    df (DataFrame): Dataset contendo a coluna.
    column (str): Nome da coluna para a qual será plotado o histograma.
  """
  plt.hist(df[column])
  plt.title(f"Histograma da coluna '{column}'")
  plt.xlabel(column)
  plt.ylabel("Frequência")
  plt.show()
```
